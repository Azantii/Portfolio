### –ü—Ä–æ–≤–µ—Ä–∫–∞ –ø—Ä–∞–≤–∏–ª—å–Ω–æ—Å—Ç–∏

```python
C = data_train['rougher.output.concentrate_au'] # –¥–æ–ª—è –∑–æ–ª–æ—Ç–∞ –≤ –∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç–µ –ø–æ—Å–ª–µ —Ñ–ª–æ—Ç–∞—Ü–∏–∏/–æ—á–∏—Å—Ç–∫–∏;
F = data_train['rougher.input.feed_au']   #  –¥–æ–ª—è –∑–æ–ª–æ—Ç–∞ –≤ —Å—ã—Ä—å–µ/–∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∞—Ç–µ –¥–æ —Ñ–ª–æ—Ç–∞—Ü–∏–∏/–æ—á–∏—Å—Ç–∫–∏;
T = data_train['rougher.output.tail_au']  #  –¥–æ–ª—è –∑–æ–ª–æ—Ç–∞ –≤ –æ—Ç–≤–∞–ª—å–Ω—ã—Ö —Ö–≤–æ—Å—Ç–∞—Ö –ø–æ—Å–ª–µ —Ñ–ª–æ—Ç–∞—Ü–∏–∏/–æ—á–∏—Å—Ç–∫–∏


data_train['recovery'] = (C * (F - T)) / (F * (C - T)) * 100 # –°–æ–∑–¥–∞—é –Ω–æ–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü —Å —Ä–∞—Å—á–∏—Ç–∞–Ω–Ω—ã–º–∏ 
                                                            # –¥–∞–Ω–Ω—ã–º–∏ —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ –æ–±–æ–≥–∞—â–µ–Ω–∏—è

mean_absolute_error(data_train['rougher.output.recovery'], data_train['recovery']) # MAE
```

–ù–∞–ø–∏—à—É —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è sMAPE –∏ –∏—Ç–æ–≥–æ–≤–æ–π sMAPE

```python
def smape(target, predictions):
    return np.mean(abs(target - predictions) / ((abs(target) + abs(predictions)) / 2))*100



def smape_fin(rough, final):
    return 0.25 * rough + 0.75 * final
```

–†–∞–∑–¥–µ–ª–∏–º –æ–±—É—á–∞—é—â—É—é –∏ —Ç–µ—Å—Ç–æ–≤—ã–µ –≤—ã–±–æ—Ä–∫–∏ –Ω–∞ –ø—Ä–∏–∑–Ω–∞–∫–∏ –∏ —Ü–µ–ª–µ–≤—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏

```python
def get_cv_scores(model, features, target, cv=5):
    my_scorer = make_scorer(smape)
    scores = cross_val_score(model, features, target, cv=5, scoring=my_scorer)
    
    return scores.mean()
```
