Краткая история ESRB

Развитие видеоигр в 1990-х годах привело к резкому увеличению графических и звуковых возможностей, а также возможности использования полноэкранного видео (FMV) в играх. В Сенате Соединенных Штатов сенаторы-демократы Джо Либерман из Коннектикута и Херб Кол из Висконсина возглавили слушания о насилии в видеоиграх и коррупции в обществе, которые начались в 1992 году. Две игры той эпохи были специально упомянуты на слушаниях из-за их содержания; файтинг Mortal Kombat содержал реалистичные оцифрованные спрайты живых актеров, кровь и возможность использовать жестокие "смертельные" приемы для победы над противниками, в то время как Night Trap содержал 90 минут контента FMV со сценами, которыесчитались сексуально вызывающими и эксплуатирующими.[1] [7] У Nintendo и Sega были разные взгляды на нежелательный контент в видеоиграх; порт Mortal Kombat для Super NES подвергся цензуре, чтобы удалить чрезмерно жестокий контент игры, в то время как порт для консолей Sega сохранил большую часть этого контента, что помогло увеличить продажи.[4][8] В мае 1993 года британские цензоры запретилиNight Trap запретили продавать в Соединенном Королевстве детям младше 15 лет, что повлияло на решение Sega создать систему возрастного рейтинга.[9]

На момент слушаний 1993 года не существовало общепромышленного стандарта для оценки видеоигр, что стало предметом разногласий на слушаниях.[10] Sega внедрила свою собственную систему добровольных рейтингов, Совет по рейтингу видеоигр (VRC), в основном для оценки игр, выпущенных для собственных консолей, что Nintendo в значительной степени оспаривала.[11] Интерактивная многопользовательская платформа 3DO имела свои собственные возрастные рейтинги, добровольно определяемые издателями игр, [12] а для оценки компьютерных игр был создан Консультативный совет по развлекательному программному обеспечению (RSAC), который использовал систему, которая оценивала интенсивность определенных классов нежелательного контента, но не использовала рекомендации по возрасту. Однако Либерман не считал, что этих систем достаточно, и в феврале 1994 года пригрозил предложить создать федеральную комиссию по регулированию и оценке видеоигр.[7] Такие магазины, как Toys "R" Us, отказались продавать игры, которые, по их мнению, были слишком жестокими для детей после слушаний.[13]

В связи с угрозой федеральных правил группа крупных разработчиков и издателей видеоигр, включая Acclaim Entertainment и Electronic Arts, а также Nintendo и Sega, в апреле 1994 года сформировала политическую торговую группу, известную как Ассоциация интерактивного цифрового программного обеспечения, с целью создания саморегулируемой основы для оценки и оценки видеоигр.игры. В то время как Sega предложила индустрии использовать свою рейтинговую систему VRC, представители Nintendo возражали против этой идеи, поскольку не хотели ассоциировать себя с работой своего главного конкурента; вместо этого была разработана независимая от поставщика рейтинговая система, известная как Рейтинговая доска развлекательного программного обеспечения (ESRB). О создании ESRB было официально объявлено Конгрессу 29 июля 1994 года. ESRB был официально запущен 16 сентября 1994 года; его система состояла из пяти возрастных рейтингов: "Раннее детство", "От детей до взрослых" (позже переименован в "Все" в 1998 году), "Подросток", "Зрелый" и "Только для взрослых". ESRB была первой рейтинговой системой, которая также использовала "дескрипторы" с краткими пояснениями содержимого, содержащегося в игре, поскольку ESRB обнаружил, что родители хотят знать об этом типе контента, прежде чем покупать игры для своих детей.[11][14][10]

## Проверка гипотез

### Гипотеза первая: средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Нулевую гипотезу формируют по знаку равенства. Соответственно Н0 будет звучать как: средние пользовательские рейтинги платформ Xbox One и PC равны.
Альтернативная же то, что они различаются.
Альфу будем брать общепринятую в 5%

```python
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу

results = st.ttest_ind(
    data_4.query('platform == "XOne"')['user_score'], 
    data_4.query('platform == "PC"')['user_score'], equal_var = False, nan_policy='omit')

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Вероятность того, что среднее значение пользовательских оценок в этих выборках отличаются по случайному стечению обстоятельств достаточно высока (~14%). А потому мы не можем утверждать, что средние оценцки пользователей Xbox One отличаются от пользователей ПК.

### Средние пользовательские рейтинги жанров Action и Sports разные

Нулевую гипотезу формируют по знаку равенства. Альтернативная по не равенству или больше или меньше

Нулевая гипотез сформулируем следующим образом: средние пользовательские рейтинги жанров Action и Sports равны.
Альтернативная: средние пользовательские рейтинги жанров Action и Sports различаются
Альфа возьмем также в 5%

```python
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу

results = st.ttest_ind(
    data_4.query('genre == "Action"')['user_score'], 
    data_4.query('genre == "Sports"')['user_score'], equal_var = False, nan_policy='omit')

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Сработала альтернативная теория, то есть с большой долей вероятностью что рейтинг для жанра Sports и Action не совпадают.
